# Task 4.7

```python
# Create a new variable for the two busiest days and the two slowest days. as we can see done durning the assignment above the two busiest days have a vaule of 0 and 1. the two slowest days have a vaule of 3 and 4.
```

```python
result2 = []
for value in df_orders_products_merged['orders_day_of_week']:
    if value in [0,1]:
        result2.append("busiest days")
    elif value in [3,4]:
        result2.append("slowest days")
    else:
        result2.append("regularly busy")
```

```python
df_orders_products_merged['busiest_days'] = result2
```

```python
df_orders_products_merged.head()
```

```python
df_orders_products_merged['busiest_days'].value_counts(dropna = False)
```

```python
# checking for the busiest hours of the day
df_orders_products_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
#creating groups for periods of time of day
hours = []
for x in df_orders_products_merged['order_hour_of_day']:
    if x in range (0,6):
        hours.append("12AM-5AM")
    elif x in range (6,12):
        hours.append("6AM-11AM")
    elif x in range (12,18):
        hours.append("12PM-5PM")
    else:
        hours.append("6PM-11PM")
```

```python
df_orders_products_merged['period_of_time'] = hours
```

```python
#here we can see that the busiest time for orders is between 12pm and 5pm. the slowest is between 12am and 5am.
df_orders_products_merged['period_of_time'].value_counts(dropna = False)
```

```python
#now to create a new column with what orders are during the busiest, least busy, and averge busy time of order
```

```python
busiest_times = []
for x in df_orders_products_merged['order_hour_of_day']:
    if x in range (12,18):
        busiest_times.append("Most Orders")
    elif x in range (0,6):
        busiest_times.append("Fewest Orders")
    else:
        busiest_times.append("Average Orders")
```

```python
df_orders_products_merged['busiest_times'] = busiest_times
```

```python
#checking the counts for busiest times
df_orders_products_merged['busiest_times'].value_counts(dropna = False)
```
